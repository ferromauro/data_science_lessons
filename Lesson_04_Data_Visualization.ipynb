# Data Visualization
La Data VIsualisation serve per mostrare i risultati ottenuti dall'analisi dei dati in modo efficace ed intuitivo. A seconda del pubblico e del tipo di dati che si vogliono mostrare ci sono diverse tecniche di data visualization.


```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

## Story Telling
Viene utilizzata per rivolgersi ad un pubblico non esperto nell'analisi dei dati, come potrebbe essere un manager che deve prendere delle decisioni in bse al risultato di una determinata analisi di dati. Si tende e limitare le infornazioni per focalizzarsi sui punti chiave. Static Images o semplici Dashbord vengono impiegate per mostrare graficamente dei risultati di analisi.



## Data Showcasing 
All'opposto dello story telling troviano il data showcasing che principalmente è rivolto ad analisti, ingenieri e statistici. Contiene numerosi dati al suo interno per poter permettere a chi lo guarda di interpretare i dati ed di arrivare da soli alle proprie considerazioni. Per questo tipo di data visualization vengono impiegate immagini e dashboard interattive.

## Data Art 
Viene usata spesso nel giornalisno o da attiviti e blogger per imprimere una impressione ed incitare all'azione utilizzando in dati per mostrare un fatto o una situazione. Pricipalmente vengono utilizzare delle immagini statiche con una particolare scelta dei colori o del metodo creativo con cui vengono rappresentati in dati.

## Object-oriented method
Prima abbiamo visto il Functional Method per creare dei grafici, ora andiamo a vedere l'Oject-oriented method che consiste nel creare una figura bianca per poi andare ad aggiungere i vali elementi del grafico.
- Creiamo una blank figure
- Aggiungiamo gli assi
- Generiamo un grafido con gli object figure
- Specifichiamo il layout del grafico con uno o più grafici

#### Definire gli assi e le griglie

```python
%matplotlib inline 
rcParams['figure.figsize']= 5,4 # definisce la dimensione delle figure dei nostri grafici. 
```

```python
x = range (1,10)
y = [1,2,3,4,0,4,3,2,1]
fig = plt.figure()                 # Creiamo la figura
ax = fig.add_axes([.1,.1,1,1])     # Aggiungiamo un sistema di assi
ax.plot(x,y)                       
```

```python
# Con le funzioni set_xticjs e set_yticks possiamo scegliere quali valori mostrare sugli assi.
x = range (1,10)
y = [1,2,3,4,0,4,3,2,1]
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([3,5])
ax.set_xticks([1,2,5,6,7,9])
ax.set_yticks([1,2,3,4,5]) # sovrascrive set_ylim()

ax.plot(x,y)

```

```python
# La funzione grid() agginge una griglia al nostrog grafico.
x = range (1,10)
y = [1,2,3,4,0,4,3,2,1]
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.grid()
ax.plot(x,y)
```

## Subplots
Una figura può contenere al suo interno più assi (quindi oiù grafici). 
Con plt.sublot(r,c) andiamo a definire come vengono disposti gli assi, in particolare: 
- **r** - corrisponderà al numero di righe  
- **c** - corrisponderà al numero di colonne.


```python
fig = plt.figure()
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x)
ax2.plot(x,y)
```

#### Lines
Matplotlib ha una serie di stili per la visualizzazzione delle linee:
- '-' linea continua
- '--' linea tratteggiata
- '-.' serie di punto e linea
- ':' serie di punti


```python
x = range(1,10)
y = [9,5,8,6,5,8,7,8,9]
x1 = range(1,10)
y1 = [9,8,7,6,5,4,3,2,1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.plot(x,y , ls=":")
plt.plot(x1,y1, ls="--", lw=5)
```


#### Markers

#### Object-oriented method

```python
rcParams["figure.figsize"]= (10,6)
address = "./Data/mtcars.csv"
cars = pd.read_csv(address)
cars.columns= ["car_names","mpg","cyl","disp","hp","drat", "wt","qsec", "vs", "am","gear", "carb"]
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,.5])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize="medium")
ax.set_xlabel("Cars Names")
ax.set_ylabel("Miles per Gallons")
ax.set_title("Miles per gallos for cars in mcars DataSet")
plt.grid()
plt.legend(loc="best")
plt.show()
```

#### Annotation
Per aggiungere una nota viene usato il metodo .annotation(xy, xytext, arroeprop)
Dove:
- xy è la posizione del punto a cui fa riferimento l'annotazione
- xytext è la posizione del testo
- arrowprop disegna una freccia dal testo al punto dell'annotazione

```python
maximum = cars.loc[cars.mpg==mpg.max()]
```

```python
rcParams["figure.figsize"]= (10,10)
address = "./Data/mtcars.csv"
cars = pd.read_csv(address)
cars.columns= ["car_names","mpg","cyl","disp","hp","drat", "wt","qsec", "vs", "am","gear", "carb"]
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,.5])
mpg.plot()
ax.set_xlabel("Cars Names")
ax.set_ylabel("Miles per Gallons")
ax.set_title("Miles per gallons for cars in mcars DataSet")
ax.annotate("Toyota Corolla", xy=(19,maximum["mpg"]),xytext=(21,35),arrowprops=dict(facecolor='red', shrink=0.05))  
ax.set_ylim([0,45])
plt.show()
```
