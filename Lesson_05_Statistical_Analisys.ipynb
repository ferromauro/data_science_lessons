# Statistical Analisys
L'anasisi statistica dei dati con python e matplolib permette di:
- individuare gli outliers
- visualizzare la distribuzione dei dati
- dedurre i tipi di variabile
- individuare correlazioni tra diverse variabili del dataset

```python
import seaborn as sb
sb.set_style("whitegrid")

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from numpy.random import randn

import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams["figure.figsize"] = (5,4)
```

```python
address = "./Data/mtcars.csv"
df = pd.read_csv(address)
df = df.rename(columns={"Unnamed: 0":"car_names"})
df.head()
```

### Visualizzare le distribuzioni dei dati con gli istogrammi

```python
# Visualizzazione di un istogramma con la funzione plot di pandas
df.index = df.car_names
mpg = df["mpg"]
mpg.plot(kind="hist")
```

```python
# Visualizzazione di un istogramma con la funzione hist di matplot
plt.hist(mpg)
plt.plot()

# praticamente Ã¨ lo stesso risultato!
```

```python
# Visualizzazzione con seaborn e la sua funzione distplot()
sb.distplot(mpg)
```

### Scatterplot


```python
df.plot(kind="scatter", x ="hp", y ="mpg", c=["orange"], s=50 )
```

```python
sb.regplot(x="hp",y="mpg", data=df, scatter=True)
```

### Pairplot (scatterplot matrix)
La funzione pairplot() ci rende una griglia di scatterplot per visualizzare eventuali correlazioni tra le coppie in un set di dati. La funzione pairplot crea una griglia di assi in modo tale che ogni variabile nei dati venga condivisa nell'asse y su una singola riga e nell'asse x su una singola colonna. CiÃ² crea grafici come mostrato di seguito.


```python
# La funzione pairplot ci permette di avere una griglia per individuare le relazioni tra coppie di misurazioni.  
sb.pairplot(df)
```

```python
# Una volta individuate eventuali correlazioni possiamo andare a creare un subset.
df_subset = df[["mpg","disp", "wt"]]
sb.pairplot(df_subset)
```

Con la funzione di Seaborn pairplot() Ã¨ possibile andare ad individuare facilmente la presenza di una correlazione tra due elementi del dataset. Possiamo anche andare ad individuare il tipo di distribuzione e se sono presenti degli outiers.

### Boxplot
Con il boxplot possiamo andare a visualizzare i classici indicatori di dispersione come:
- **campo di variazione**: intervallo tra valore minimo e valore massimo;
- **mediana**: il valore che si trova nel mezzo della distribuzione;
- **scarto interquartile**: range di valori che sono contenuti tra il primo e terzo quartile, che include la metÃ  *'centrale'* delle misurazioni;

```python
df.boxplot("mpg", by="am")
df.boxplot("wt", by="am")
```

```python
# Boxplot realizzato con la libreria Seaborn.
rcParams["figure.figsize"]=(5,10)
sb.boxplot(x="am", y="mpg", data=df, palette="Accent")
```
