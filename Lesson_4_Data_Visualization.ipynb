# Data Visualization
La Data VIsualisation serve per mostrare i risultati ottenuti dall'analisi dei dati in modo efficace ed intuitivo. A seconda del pubblico e del tipo di dati che si vogliono mostrare ci sono diverse tecniche di data visualization.


```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

## Story Telling
Viene utilizzata per rivolgersi ad un pubblico non esperto nell'analisi dei dati, come potrebbe essere un manager che deve prendere delle decisioni in bse al risultato di una determinata analisi di dati. Si tende e limitare le infornazioni per focalizzarsi sui punti chiave. Static Images o semplici Dashbord vengono impiegate per mostrare graficamente dei risultati di analisi.



## Data Showcasing 
All'opposto dello story telling troviano il data showcasing che principalmente Ã¨ rivolto ad analisti, ingenieri e statistici. Contiene numerosi dati al suo interno per poter permettere a chi lo guarda di interpretare i dati ed di arrivare da soli alle proprie considerazioni. Per questo tipo di data visualization vengono impiegate immagini e dashboard interattive.

## Data Art 
Viene usata spesso nel giornalisno o da attiviti e blogger per imprimere una impressione ed incitare all'azione utilizzando in dati per mostrare un fatto o una situazione. Pricipalmente vengono utilizzare delle immagini statiche con una particolare scelta dei colori o del metodo creativo con cui vengono rappresentati in dati.

### Line Chart
Viene utilizzato per mostrare la variazione dei dati nel tempo.

#### Creare un grafico da una serie di valori


```python
x= range(1,10)
y= [1,4,2,5,7,3,2,6,3]
```

```python
 plt.plot(x,y)
```

#### Creare un grafico da un DataFrame


```python
address = "./Data/mtcars.csv"
cars = pd.read_csv(address)
cars.columns= ["car_names","mpg","cyl","disp","hp","drat", "wt","qsec", "vs", "am","gear", "carb"]
```

```python
cars.head()
```

```python
mpg = cars['mpg']
mpg.plot()
```

#### Visualizzare piÃ¹ colonne di un DataFrame


```python
df = cars[['cyl','wt', 'mpg']]
df.plot()
```

### Bar Chart
Viene utilizzato per mostrare la variazione dei dati considerando diverse categorie all'interno del dataset.


#### VIsualizzare dati da una serie di valori

```python
plt.bar(x,y)
```

#### Visualizzare i dati di un DataFrame

```python
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

### Pie Chart
Serve per mostrare l'incidenza di diverse categorie di dati all'interno dell'intero dataset.

```python
x= [3,4,3,6,4,6]
plt.pie(x)
plt.show()
```
